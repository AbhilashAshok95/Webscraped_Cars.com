# Pagitation - Multiple Page

```python
name = []
mileage = []
dealer_name = []
rating = []
review_count = []
price = []

for i in range(1,50):
    website = 'https://www.cars.com/shopping/results/?page='+str(i)+ '&page_size=20&list_price_max=&makes[]=mercedes_benz&maximum_distance=20&models[]=&stock_type=all&zip='
    
    #requesting permision
    response = requests.get(website)
    
    #soup
    soup = BeautifulSoup(response.content,'html.parser')
    
    #results
    results = soup.find_all('div',{'class':'vehicle-card'})
    
    
    
    
    for result in results:

        #name
        try:
            name.append(result.find('h2').get_text())
        except:
            name.append('NA')

         #mileage
        try:
            mileage.append(result.find('div',{'class':'mileage'}).get_text())
        except:
            mileage.append('NA')   

         #dealer_name
        try:
            dealer_name.append(result.find('div',{'class':'dealer-name'}).get_text())
        except:
            dealer_name.append('NA')  

         #rating
        try:
            rating.append(result.find('span',{'class':'sds-rating__count'}).get_text())
        except:
            rating.append('NA')  

        #review Count
        try:
            review_count.append(result.find('span',{'class':'sds-rating__link sds-button-link'}).get_text())
        except:
            review_count.append('NA') 

        #Price
        try:
            price.append(result.find('span',{'class':'primary-price'}).get_text())
        except:
            price.append('NA')
    
    
    
    
```

```python
car_dealer = pd.DataFrame({'Name':name,'mileage':mileage,'Dealer_name': dealer_name ,'Rating':rating,
                           'Review_count':review_count,'Price':price})

car_dealer['Review_count'] = car_dealer['Review_count'].apply(lambda x:x.strip('reviews)').strip('('))

car_dealer['Dealer_name'] = car_dealer['Dealer_name'].apply(lambda x:x.strip('\n'))
```

```python
car_dealer
```

```python
split = car_dealer['Name'].str.split(' ',n=3, expand = True)
split
```

```python
final = pd.concat([car_dealer, split], axis=1, join="inner")
final
```

```python
final.to_excel('Mercedez_new_all.xlsx', index=True)
```

```python
car_dealer = pd.DataFrame({'Name':name,'mileage':mileage,'Dealer_name': dealer_name ,'Rating':rating,
                           'Review_count':review_count,'Price':price})

car_dealer['Review_count'] = car_dealer['Review_count'].apply(lambda x:x.strip('reviews)').strip('('))

car_dealer['Dealer_name'] = car_dealer['Dealer_name'].apply(lambda x:x.strip('\n'))

car_dealer
```

```python
split = car_dealer['Name'].str.split(' ',n=3, expand = True)
split
```

```python
final = pd.concat([car_dealer, split], axis=1, join="inner")
final
```

```python
final.to_excel('Toyota_all.xlsx', index=True)
```
